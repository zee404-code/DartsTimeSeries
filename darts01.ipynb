```python
#from darts.datasets import AirPassengersDataset
from darts.datasets import HeartRateDataset
series = HeartRateDataset().load()
series.plot()
#series.plot(linewidth = 1)
#set_figheight = 50000
#set_figwidth = 10000000
```

```python
#timeseries operations 
#1 splitting
training, validation = series.split_before(0.88)
training.plot()
validation.plot()
```

```python
#2 slicing
series1, series2 = series[:36], series[36:] #+36 is from the left. -36 is from the right
series1.plot()
series2.plot()
```

```python
series.diff().plot() #differencing
```

```python
series.diff().diff().plot() #can do multiple differencing
```

```python
from darts.models import ARIMA
model = ARIMA()
print(model)
model.fit(training)
pred = model.predict(n=60)
pred.plot(label='pred')
#training.plot(label='train')
```
